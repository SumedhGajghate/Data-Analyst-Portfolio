Step 1- Data preperation/cleaning
Tools- Pandas and Jupyter Notebook

```python
#1.1-Load the data
import pandas as pd

df = pd.read_csv('customer_shopping_data.csv')
```

```python
#1.2-Inspect the data

df.head(3)
```

```python
#basic information about the dataset

df.info()
```

```python
#summary statistics

df.describe()
```

```python
#1.3- Handle missing value 

df.isnull().sum()
```

```python
#1.4- Change the data-type for date colunm

df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

```

```python
#1.5 Remove duplicates 

df.duplicated().sum()
```

```python
#1.6 Creaating new feature/ column 

#Customer segmentation based on age group

bins = [18, 24, 34, 44, 54, 64, 70]
la = ['Young Adults','Young Professionals','Established Professionals','Middle-aged Adults','Pre-retirees','Early Retirees']

df['age_based_segmentation']= pd.cut(df['age'], bins=bins, labels=la, right=False)
```

```python
#Customer segmentation based on visits...

df['count_visit'] = df.groupby('customer_id')['invoice_no'].transform('count')

def visits(v):
    if v >10:
        return 'Regular'
    elif 3<= v <=10:
        return 'Semi-Regular'
    else: return 'New'

df['customer_type']= df['count_visit'].apply(visits)    
```

```python
# Creating Total price column
df['total_price'] = df['price'] * df['quantity']
```

```python
df = df.drop('totla_price', axis=1 )
df.head()
```

```python
#1.7- Removing outliers

import numpy as np

Q1= df['price'].quantile(.25)
Q3= df['price'].quantile(.75)
IQR = Q3-Q1

lower = Q1 - 1.5*IQR
uper = Q3 + 1.5*IQR

df = df[ (df['price'] >=lower)  & (df['price'] <=uper)]
```

```python
#1.8 -Saving cleaned data 
df.to_csv('cleaned_customer_shopping_dataset.csv', index=False)
```

```python
# 2.3 - Bivariate Analysis

#Numirical vs numerical type
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('cleaned_customer_shopping_dataset.csv')

#on price and quantity
plt.figure(figsize=(5,3))
plt.scatter(df['price'], df['quantity'])
plt.xlabel('price')
plt.ylabel('Quantity')
plt.show

# Inference - No signnificent correlation
```

```python
# Numirical vs Categorial type

#on price vs category

sns.boxplot(data=df, x='category', y='price')
plt.xlabel('category')
plt.ylabel('price')
plt.xticks(rotation=90)
```

```python
#categorical vs categorical

#On gender vs category
plt.figure(figsize=(11,4))
sns.countplot(data=df, x='category', hue='gender')
plt.xlabel('category')
plt.ylabel('Count')
plt.show
```

```python
#categorical vs categorical

#on gender and payment method

cross_tab = pd.crosstab(df['gender'], df['payment_method'])

sns.heatmap(data = cross_tab, annot=True, cmap='Blues')
plt.xlabel('Gender')
plt.ylabel('payment_method')
plt.show
```

```python
#2.4 - Time Series Analysis

#Trend analysis - Line Plot (Mounth wise)
df['invoice_date'] = pd.to_datetime(df['invoice_date'])

sales_trend = df.groupby(pd.Grouper(key='invoice_date', freq='ME'))['total_price'].sum()
sales_trend.plot()
plt.xlabel('Date')
plt.ylabel('Total sales')
plt.show()
```

```python
# Seasonality:
# Check for patterns that repeat over time, such as monthly or weekly trends.

df['month'] = df['invoice_date'].dt.month
df['day_of_week'] = df['invoice_date'].dt.dayofweek

month_group = df.groupby('month')['total_price'].sum()

month_group.plot(kind= 'bar')
plt.title('Weekly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(range(12), [month.strftime('%B') for month in pd.date_range(start='2021-01-01', periods=12, freq='ME')])
plt.show()
```

```python
day_group = df.groupby('day_of_week')['total_price'].sum()
day_group.plot(kind= 'bar')
plt.title('Weekly Sales')
plt.xlabel('Days of Week')
plt.ylabel('Sales')
plt.show()
```

```python
#2.5 - Multivariate Analysis
#Analyze relationships between multiple variables:

sns.pairplot(df[['age', 'quantity', 'price', 'total_price']])
plt.show()
```
